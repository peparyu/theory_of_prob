# Нормальное распределение

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
plt.plot(x, stats.norm.pdf(x, mu, sigma), color='blue')
plt.show()
```

```python
norm_rv = norm(loc=0, scale=1)
sample = norm_rv.rvs(1000)

n2 = 1
sample_2_mean = []
for k in range(1000):
    sample_2 = np.random.choice(sample, n2)
    sample_2_mean.append(np.mean(sample_2))
sat_score = pd.DataFrame({"score": sample_2_mean})
sat_score.head()
```

```python
mu = 0 
sigma = 1

x_left = -1.75
x_right = 0.25




fig, ax = plt.subplots(figsize=(10, 6))

# Plot the pdf of the normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
ax.plot(x, stats.norm.pdf(x, mu, sigma), color="#4570bf", label="Нормальное распределение")
# ax.plot(density["x"], density["pdf"], color="#4570bf", label="Нормальное распределение")

# Draw a red vertical line at x = 800
# ax.axvline(x=-1, color="darkblue", linestyle="--", label="x = -1")
# btw0_1 = norm.cdf(1, mu, sigma) - norm.cdf(0, mu, sigma)
ax.axvline(x=x_left, ymin=0., ymax=1, color="darkblue", linestyle="-.")
ax.axvline(x=x_right,  ymin=0., ymax=1, color="darkblue", linestyle="-.")


area_under_curve = (norm.cdf(x_right, mu, sigma) - norm.cdf(x_left, mu, sigma)) 
x1 = np.linspace(x_left,x_right)
ax.fill_between(x1, norm.pdf(x1, mu, sigma), color="#F6AD55", alpha=1, label=f"{area_under_curve * 100:.2f}%")
ax.text(-0.8, 0.00015, f"{area_under_curve * 100:.2f}%", color="black", fontsize=30, ha='center')


ax.set(xlabel=r"$X$", ylabel=r"$e^{-\frac{x^2}{2}}$")
ax.xaxis.label.set_color("#4570bf")  
ax.yaxis.label.set_color("#4570bf") 
ax.xaxis.label.set_fontsize(25) 
ax.yaxis.label.set_fontsize(25) 
ax.set_title("ПЛОТНОСТЬ НОРМАЛЬНОГО РАСПРЕДЕЛЕНИЯ")

# Add the percentage value on the plot


ax.legend()

plt.xticks(color='black', fontsize=20)
plt.yticks(color='black',fontsize=20)

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.savefig('/Users/ruslanpepa/Documents/МГУ/Рукописи/ТВиМС/Графики/139a', bbox_inches='tight')
plt.show()
```
