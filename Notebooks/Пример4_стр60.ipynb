```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import pandas as pd
import math
```

```python
plt.style.use("ggplot")
plt.rcParams.update({
    "figure.figsize": (8, 5),
    "figure.dpi": 100,
    "savefig.dpi": 300,
    "figure.constrained_layout.use": True,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "legend.fontsize": 9,
    "legend.title_fontsize": 10,
        
})
```

```python
N = 10000
density = pd.DataFrame()
density["x"] = np.linspace(-1, 1 + 0.01, N)
density["1-|x|"] = 2*np.abs(density["x"]) 
density["0"] = np.full(N,0)
# Calculate the area under the curve up to x = -1

# area_under_curve = expon.cdf(1, mu, sigma)

fig, ax = plt.subplots(figsize=(10, 6))
x = np.linspace(1,np.exp(1),N)
y = x*0
# Plot the pdf of the normal distribution
ax.plot(x,  np.log(x), color="darkblue",linewidth=3)
ax.plot(np.linspace(0.5,1,N),y , color="darkblue",linewidth=3 )
ax.plot(np.linspace(np.exp(1),3,N),np.full(N,0) , color="darkblue", label="f(x)=log(x)",linewidth=3)
# ax.plot(np.full(N,0),np.linspace(0,1,N) , color="green",label='')
# ax.plot(np.linspace(-1,1,N),np.full(N,0), color="red",label='')


# Fill the specified region with red color
ax.fill_between(x,  np.log(x), color="blue", alpha=0.1, label=" ")

ax.set(xlabel="X")

# ax.xaxis.label.set_color("darkblue")  
# ax.yaxis.label.set_color("darkblue")  

ax.set_title("  ", color="darkblue")

# Add the percentage value on the plot
# ax.text(0.8, 0.64, f"{area_under_curve * 100:.2f}%", color="red", fontsize=10, ha='center')

ax.legend()

plt.xticks(color='black')
plt.yticks(color='black')

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)
ax.margins(-0.03, 0.03)

# Display the plot
plt.savefig('/Users/ruslanpepa/CoveringFlow/den_abs.pdf', bbox_inches='tight')
plt.show()
```
