```python
plt.style.use("fivethirtyeight")
# Enable LaTeX rendering
# plt.rcParams['text.usetex'] = True
# plt.rcParams['font.family'] = 'serif'

# Data of the problem
fig, ax = plt.subplots(figsize=(10, 6))
x_min = 1 #   
x_max = 3 #   
ro = lambda x: 1 - x/4 #   
F_x = lambda x: x - x**2/8 - 7/8 #   
x_left = 1
x_right = 2
delta_step = 3 #       

# Build the graphs
x1 = np.linspace (x_min - delta_step, x_min , 100)
y1 = [0]*100
x2 = np.linspace(x_min, x_max, 100)
y2 = ro(x2)
x3 = np.linspace (x_max, x_max + delta_step, 100)
y3 = [0]*100
ax.plot(x1, y1, color='#4570bf')
ax.plot(x2, y2, color='#4570bf')
ax.plot(x3, y3, color='#4570bf')


# Fill the specified region with red color

x4 = np.linspace(x_left, x_right, 100)
y4 = ro(x4)
ax.fill_between(x4, y4 , color="#F6AD55", alpha=0.5, label="")
# area under the curve
area_under_curve = (F_x(x_right) - F_x(x_left))
# Add the percentage value on the plot
ax.text(x_right+1, ro(x_right)-0.1, f"{area_under_curve * 100:.2f}%", color="#F6AD55", fontsize=30, ha='center')
ax.axvline(x=x_left,  ymin=0., ymax=ro(-1), color="#4570bf",linewidth=2, linestyle="-")
ax.axvline(x=x_right,  ymin=0., ymax=ro(x_right), color="#4570bf",linewidth=2, linestyle="-")




# Customize the name
ax.set_title("  ",color = '#A0AEC0', fontsize=20)

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)
ax.margins(-0.03, 0.03)

# Customize the label
ax.set(xlabel="$x$")
ax.set(ylabel=r"$\rho(x) =1- \frac{x}{4}$")
# ax.set_xlabel(r'\textbf{Time (s)}')
# ax.set_ylabel(r'$\rho(x)$')

# Customize the axis
ax.xaxis.label.set_color("#4570bf")  
ax.yaxis.label.set_color("#4570bf") 
ax.xaxis.label.set_fontsize(25) 
ax.yaxis.label.set_fontsize(25) 
# Customize the tick
plt.xticks(color='black', fontsize=20)
plt.yticks(color='black',fontsize=20)
#Customize the legend
ax.legend(fontsize=20)




plt.savefig('/Users/ruslanpepa/Documents/////121density.png', bbox_inches='tight')
```
