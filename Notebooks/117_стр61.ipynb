```python
plt.style.use("ggplot")
plt.rcParams.update({
    "figure.figsize": (8, 5),
    "figure.dpi": 100,
    "savefig.dpi": 300,
    "figure.constrained_layout.use": True,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "legend.fontsize": 9,
    "legend.title_fontsize": 10,
        
})
```

```python
plt.style.use("fivethirtyeight")
#   
fig, ax = plt.subplots(figsize=(10, 6))
x_min = -3
x_max = 2
ro = lambda x: x**4/55
F_x = lambda x: ((x**5 + 243)/275)


x1 = np.linspace (x_min - 2, x_min , 100)
y1 = [0]*100
x2 = np.linspace(x_min, x_max, 100)
y2 = ro(x2)
x3 = np.linspace (x_max, x_max + 2, 100)
y3 = [0]*100
ax.plot(x1, y1, color='blue')
ax.plot(x2, y2, color='blue')
ax.plot(x3, y3, color='blue')
ax.plot(np.full(100,2),np.linspace(0,ro(2),100) , color="blue")
ax.plot(np.full(100,-3),np.linspace(0,ro(-3),100) , color="blue")
# Fill the specified region with red color

x4 = np.linspace(-2, 2, 100)
y4 = ((x4**4)/55)
ax.fill_between(x4, y4 , color="red", alpha=0.5, label="")

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)
ax.margins(-0.03, 0.03)


ax.set(xlabel="x")
ax.set(ylabel="ro(x)")

ax.xaxis.label.set_color("darkblue")  
ax.yaxis.label.set_color("darkblue") 
ax.xaxis.label.set_fontsize(35) 
ax.yaxis.label.set_fontsize(35) 

plt.xticks(color='black', fontsize=30)
plt.yticks(color='black',fontsize=30)

ax.set_title("   ", color="darkblue",fontsize=30)

# area under the curve
area_under_curve = (F_x(2) - F_x(-2))

# Add the percentage value on the plot
ax.text(0.8, 0.1, f"{area_under_curve * 100:.2f}%", color="red", fontsize=30, ha='center')
plt.savefig('/Users/ruslanpepa/Documents/////117density.png', bbox_inches='tight')

ax.legend(fontsize=20)
```
