# Нормальное распределение

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
import scipy.stats as stats
import math
import pandas as pd
import numpy as np
plt.style.use("ggplot")
plt.rcParams.update({
    "figure.figsize": (8, 5),
    "figure.dpi": 100,
    "savefig.dpi": 300,
    "figure.constrained_layout.use": True,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "legend.fontsize": 9,
    "legend.title_fontsize": 10,
        
})
```

```python
mu = 7
sigma = 5

area = 6

x_left = 9
x_right = area*sigma
p_ = 0.05 # Квантиль нужного уровня



fig, ax = plt.subplots(figsize=(10, 6))

# Plot the pdf of the normal distribution
x = np.linspace(mu - area*sigma, mu + area*sigma, 1000)
ax.plot(x, stats.norm.pdf(x, mu, sigma), color="#4570bf")
quantile = stats.norm.ppf(p_, loc=mu, scale=sigma)

# Draw a red vertical line at x = 800
# ax.axvline(x=-1, color="darkblue", linestyle="--", label="x = -1")
# btw0_1 = norm.cdf(1, mu, sigma) - norm.cdf(0, mu, sigma)
ax.axvline(x=x_left, ymin=0., ymax=1, color="darkblue", linestyle="-.")
ax.axvline(x=x_right,  ymin=0., ymax=1, color="darkblue", linestyle="-.")
ax.axvline(x=mu,  ymin=0., ymax=1, color="red", linestyle="-",label="СРЕДНЕЕ ЗНАЧЕНИЕ")



area_under_curve = (norm.cdf(x_right, mu, sigma) - norm.cdf(x_left, mu, sigma)) 
x1 = np.linspace(x_left,x_right)
ax.fill_between(x1, norm.pdf(x1, mu, sigma), color="#F6AD55", alpha=0.5, label=f"{area_under_curve * 100:.2f}%")
ax.text(mu+4*sigma, 0.00015, f"{area_under_curve * 100:.2f}%", color="#F6AD55", fontsize=30, ha='center')

ax.axvline(x=quantile,  ymin=0., ymax=1, color="green", linestyle="-.",label=f"Q_{p_}={quantile:.2f}")
x2 = np.linspace(mu-4*sigma,quantile)
area_under_curve = (norm.cdf(quantile, mu, sigma) - norm.cdf(mu-4*sigma, mu, sigma))
ax.fill_between(x2, norm.pdf(x2, mu, sigma), color="green", alpha=0.3, label=f"{area_under_curve * 100:.2f}%")

ax.set(xlabel=r"$X$", ylabel=r"$e^{-\frac{x^2}{2}}$")
ax.xaxis.label.set_color("#4570bf")  
ax.yaxis.label.set_color("#4570bf") 
ax.xaxis.label.set_fontsize(25) 
ax.yaxis.label.set_fontsize(25) 
ax.set_title("ПЛОТНОСТЬ НОРМАЛЬНОГО РАСПРЕДЕЛЕНИЯ")

# Add the percentage value on the plot


ax.legend(fontsize=15)

plt.xticks(color='black', fontsize=20)
plt.yticks(color='black',fontsize=20)

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.savefig('/Users/ruslanpepa/Documents/МГУ/Рукописи/ТВиМС/Графики/142', bbox_inches='tight')
plt.show()
```
