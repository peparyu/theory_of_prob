```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
from scipy.stats import expon
import seaborn as sns
import pandas as pd
```

```python
plt.style.use("ggplot")
plt.rcParams.update({
    "figure.figsize": (8, 5),
    "figure.dpi": 100,
    "savefig.dpi": 300,
    "figure.constrained_layout.use": True,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "legend.fontsize": 9,
    "legend.title_fontsize": 10,
        
})
```

```python
# defining theme
sns.set(style="white")
colors = sns.color_palette("viridis", n_colors=3)

# plotting the SAT score distribution using Seaborn with a KDE (Kernel Density Estimate) plot
plt.figure(figsize=(10, 6))

mu = 0 
sigma = 1

# norm.cdf() calculates the cumulative probability
prob_less_than_minus_one = expon.cdf(1, mu, sigma)
print(
    "The probability that a student will score less than 1",
    round(prob_less_than_minus_one, 4),
)
```

```python
mu = 0 
sigma = 1

# norm.cdf() calculates the cumulative probability
prob_less_than_one = expon.cdf(1, mu, sigma)
print(
    "The probability that a student will score less than 1",
    round(prob_less_than_one, 4),
)

density = pd.DataFrame()
density["x"] = np.linspace(
    -1, 5 + 0.01, 100
)
density["pdf"] = expon.pdf(density["x"], mu, sigma)
# Calculate the area under the curve up to x = -1

area_under_curve = expon.cdf(1, mu, sigma)

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the pdf of the normal distribution
ax.plot(density["x"], density["pdf"], color="red", label=" ")

# Draw a red vertical line at x = 800
ax.axvline(x=1, color="darkblue", linestyle="--", label="x = 1")

# Create an array of 50 numbers between min SAT score and 800
x1 = np.linspace(density["x"].min(), 1)

# Fill the specified region with red color
ax.fill_between(x1, expon.pdf(x1, mu, sigma), color="red", alpha=0.3, label="  ")

ax.set(xlabel="X", ylabel="")

ax.xaxis.label.set_color("darkgreen")  
ax.yaxis.label.set_color("darkgreen")  

ax.set_title(" ", color="darkgreen")

# Add the percentage value on the plot
ax.text(1.25, 0.00015, f"{area_under_curve * 100:.2f}%", color="darkblue", fontsize=10, ha='center')

ax.legend()

plt.xticks(color='orange')
plt.yticks(color='orange')

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.savefig('/Users/ruslanpepa/CoveringFlow/den_of_expon_dist.png', bbox_inches='tight')
plt.show()
```
