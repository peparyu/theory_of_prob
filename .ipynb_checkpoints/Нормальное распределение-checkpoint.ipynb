```python
norm_rv = norm(loc=0, scale=1)
sample = norm_rv.rvs(1000)

n2 = 1
sample_2_mean = []
for k in range(1000):
    sample_2 = np.random.choice(sample, n2)
    sample_2_mean.append(np.mean(sample_2))

plt.hist(sample_2_mean, bins = 50, density = True)
```

```python
mu = 0 
sigma = 1

# norm.cdf() calculates the cumulative probability
prob_less_than_minus_one = norm.cdf(-1, mu, sigma)
# print(
#     "The probability that a student will score less than -1",
#     round(prob_less_than_minus_one, 4),
# )

density = pd.DataFrame()
density["x"] = np.linspace(
    sat_score["score"].min() - 0.01, sat_score["score"].max() + 0.01, 100
)
density["pdf"] = norm.pdf(density["x"], mu, sigma)

# Calculate the area under the curve up to x = 800

area_under_curve = norm.cdf(-1, 0, 1)

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the pdf of the normal distribution
ax.plot(density["x"], density["pdf"], color="red", label="Нормальное распределение")

# Draw a red vertical line at x = 800
# ax.axvline(x=-1, color="darkblue", linestyle="--", label="x = -1")
# btw0_1 = norm.cdf(1, mu, sigma) - norm.cdf(0, mu, sigma)
ax.axvline(x=-1, ymin=0., ymax=0.6, color="darkblue", linestyle="--")
ax.axvline(x=1,  ymin=0., ymax=0.6, color="darkblue", linestyle="--")
# Create an array of 50 numbers between min SAT score and 800
# x1 = np.linspace(density["x"].min(), -1)
x1 = np.linspace(-3,-2)
x2 = np.linspace(-2,-1)
x3 = np.linspace(-1,1)
# x4 = np.linspace(0,1)
x5 = np.linspace(1,2)
x6 = np.linspace(2,3)
# Fill the specified region with red color
ax.fill_between(x1, norm.pdf(x1, mu, sigma), color="red", alpha=0.1, label=f"{norm.cdf(-1, 0, 1) * 100:.2f}%")
ax.fill_between(x2, norm.pdf(x2, mu, sigma), color="red", alpha=0.3, label="Площадь под графиком")
ax.fill_between(x3, norm.pdf(x3, mu, sigma), color="red", alpha=0.6, label="Площадь под графиком")
ax.fill_between(x5, norm.pdf(x5, mu, sigma), color="red", alpha=0.3, label="Площадь под графиком")
ax.fill_between(x6, norm.pdf(x6, mu, sigma), color="red", alpha=0.1, label="Площадь под графиком")

ax.set(xlabel=r"$X$", ylabel=r"$e^{-\frac{x^2}{2}}$")
ax.xaxis.label.set_color("#4570bf")  
ax.yaxis.label.set_color("#4570bf") 
ax.xaxis.label.set_fontsize(25) 
ax.yaxis.label.set_fontsize(25) 

 

ax.set_title("ПЛОТНОСТЬ НОРМАЛЬНОГО РАСПРЕДЕЛЕНИЯ", color="darkgreen")

# Add the percentage value on the plot
ax.text(-0.8, 0.00015, f"{area_under_curve * 100:.2f}%", color="darkblue", fontsize=10, ha='center')

ax.legend()

plt.xticks(color='black', fontsize=20)
plt.yticks(color='black',fontsize=20)

# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.savefig('/Users/ruslanpepa/Documents/МГУ/Рукописи/ТВиМС/Графики/den_of_nor_dist.png', bbox_inches='tight')
plt.show()
```

```python
mu = 0 
sigma = 1

# norm.cdf() calculates the cumulative probability
prob_less_than_one = norm.cdf(1, mu, sigma)
print(
    "The probability that a student will score less than 1",
    round(prob_less_than_one, 4),
)

density = pd.DataFrame()
density["x"] = np.linspace(
    -4, 5 + 0.01, 100
)
density["pdf"] = norm.cdf(density["x"], mu, sigma)
# Calculate the area under the curve up to x = -1

area_under_curve = norm.cdf(1, mu, sigma)

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the pdf of the normal distribution
ax.plot(density["x"], density["pdf"], color="red", label="Нормальное распределение")

# Draw a red vertical line at x = 800
ax.axhline(y=prob_less_than_one, color="darkblue", linestyle="--", label="y = 1")

# Create an array of 50 numbers between min SAT score and 800
x1 = np.linspace(density["x"].min(), 1)


# ax.set(xlabel="X", ylabel="Вероятность")
# ax.xaxis.label.set_color("darkgreen")  
# ax.yaxis.label.set_color("darkgreen")  
# ax.set_title("Нормальное распределение", color="darkgreen")

# ax.plot(x, y, label=r'$\sin(x)$')
ax.set_xlabel(r' $x$ ', fontsize=20)
ax.set_ylabel(r'$e^{\frac{-x^2}{2}}$',fontsize=20)
ax.set_title(r'ФУНКЦИЯ РАСПРЕДЕЛЕНИЯ ВЕРОЯТНОСТИ ')



# Add the percentage value on the plot
ax.text(1.5, 0.85, f"{area_under_curve * 100:.2f}%", color="darkblue", fontsize=10, ha='center')
ax.legend(fontsize=20)
plt.xticks(color='black', fontsize=20)
plt.yticks(color='black',fontsize=20)


# Customize the grid lines
ax.grid(True, linestyle='--', alpha=0.7)
plt.scatter(1, prob_less_than_one, color='red', s=50)

# Display the plot
plt.savefig('/Users/ruslanpepa/Documents/МГУ/Рукописи/ТВиМС/Графики/funct_of_nor_dist.png', bbox_inches='tight')
plt.show()
```
